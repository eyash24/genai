# Messages

- fundamental unit of context in LangChain.
- represent input + output of models carrying content and metadata 
- represent state of conversation when interacting with LLM
- They are objects that contain:
    - Role: identify the message type (user, system)
    - Content: Content of message (text, images, audio, documents)
    - Metadata: Optional fields (response information, message IDs, token usage)

Provides a standard message type for all model providers

```python
import os
from langchain.chat_models import init_chat_model

os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
model = init_chat_model('groq:qwen/qwen3-32b')
```

# Text Prompts 

strings ideals for straightforward generation tasks where there is no need to retain history 

Usage: 
- single standalone request 
- no conversational history required 
- minimal code complexity 



```python
model.invoke('What is langchain? ')
```

# Message Prompts

Pass a list of message objects

Message types 
- System message: tells models how to behave and provide context for interactions 
- human message: user inputs and interaction with the model 
- ai message: responsed generated by model ( text content, tool calls, metadata)
- tool message: output of tool calls


```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage('You are a Staff Machine Learning Developer'),
    HumanMessage('Create Roadmap for Machine Learning for freshers'),
]

response = model.invoke(messages)
response.content
```

```python
# message metadata
human_msg = HumanMessage(
    content = "Hello!",
    name = 'alice',     # optional: identify diff users
    id = 'msg_123'      # optional: unique identifier for tracing
)
```

```python
response = model.invoke([human_msg])
response.content
```

```python
response
```

```python
from langchain.messages import AIMessage, SystemMessage, HumanMessage

ai_msg = AIMessage('I would be happy to help out.')

message = [
    SystemMessage('You are a helpful assistant'),
    HumanMessage('Can you help me out?'),
    ai_msg,
    HumanMessage('Great! what is 56x92093?')
]

response = model.invoke(message)
print(response.content)
```

```python
response.usage_metadata
```

```python
from langchain.messages import AIMessage, ToolMessage, HumanMessage

ai_msg  = AIMessage(
    content = [],
    tool_call = [{
        'name': "get_weather",
        "args": {'location': 'San Francisco'},
        "id": "call_123"
    }]
)

weather_result = "sunny, 72"
tool_msg = ToolMessage(
    content=weather_result,
    tool_call_id='call_123'
)

msg = [
    HumanMessage('What is weather in California?'),
    ai_msg,
    tool_msg
]

response = model.invoke(msg)
response.content
```

```python
response.usage_metadata
```
