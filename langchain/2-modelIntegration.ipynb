# Model Integration

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY')
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
model
```

```python
# model inovacation
response = model.invoke("Hello, how are you?")
response
```

```python
response.content
```

# Streaming

Models cam stream their output content while the output is being generated. This is helpful for longer messages.

```python
for chunk in model.stream('write a 100 words paragraph on AI'):
    print(chunk.text, end="|", flush=True)
```

# Batch

Batching a collections of independent requests to a model to improve performance and reduce cost. Processing is done in parallel.

```python
responses = model.batch([
    "how to cook pasta?",
    "what is 20x756 ?"
    "when was quantum computing termed ?"
])

for response in responses:
    print(response)
```

```python
responses = model.batch(
    [
        "how to cook pasta?",
        "what is 20x756 ?"
        "when was quantum computing termed ?"
    ],
    config=[
        'max_currency': 5 # limits the number of parallel calls
    ]
)
```
