# Tools

Models request tools that perform tasks such as fetching data, searching web, running code.

They are pairings of:
1. A schema : includes name of tool, description, argument definitions (often in JSON schema)
2. Function or coroutine to execute
    

```python
import os
from langchain.chat_models import init_chat_model

os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')

model = init_chat_model('groq:qwen/qwen3-32b')
```

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    '''Get the weather at a location'''
    return f'Weather at {location} is sunny'

model_with_tools = model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke('What is the weather is boston ?')
for tool_call in response.tool_calls:
    print(f'Tool: {tool_call['name']}')
    print(f'Args: {tool_call['args']}')
```

# Tool execution loop

```python
messages = [{"role":"user", "content": "what is the weather in denver?"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

print('Response before execution tool: ', ai_msg)
# The aimsg may not contain any content from the tools. Henceforth, manually call the tools and get their content.
# Use the combined ai mssg and tool content to come to final response 

for tool_call in ai_msg.tool_calls:
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)

final_rseponse = model_with_tools.invoke(messages)
```
